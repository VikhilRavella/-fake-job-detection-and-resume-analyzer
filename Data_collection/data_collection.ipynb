```python
#Handle Missing Values
# Fill NaN with empty string for text columns
text_cols = ['company_profile', 'description', 'requirements']
for col in text_cols:
    df[col] = df[col].fillna("")
```

```python

# ðŸ”¹ Step 1: Drop rows with missing 'description' or 'requirements'
df = df.dropna(subset=['description', 'requirements'])

# ðŸ”¹ Step 2: Fill missing 'department' with placeholder
df['department'] = df['department'].fillna("Unknown")

def clean_location(loc):
    if pd.isna(loc):
        return "Unknown"
    parts = [p.strip() for p in str(loc).split(',') if p.strip()]
    return ', '.join(parts)

df['location'] = df['location'].apply(clean_location)

# ðŸ”¹ Step 4: Optional â€” fill missing company_profile or salary_range
df['company_profile'] = df['company_profile'].fillna("Not Provided")
df['salary_range'] = df['salary_range'].fillna("Not Disclosed")

# ðŸ”¹ Step 5: Reset index after cleaning
df = df.reset_index(drop=True)

# ðŸ”¹ Step 6: Preview cleaned data
print(df[['job_id', 'title', 'location', 'department', 'description']].head())
```
